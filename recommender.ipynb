```python
def remove_rows_with_empty_lists(games):
    # Create a boolean DataFrame where True indicates an empty list
    mask = games.applymap(lambda x: x == '[]')
    
    # Identify rows that contain any empty lists
    rows_with_empty_lists = mask.any(axis=1)
    
    # Filter out those rows
    df_filtered = games[~rows_with_empty_lists]
    
    return df_filtered

# Remove rows with empty lists
games_data = remove_rows_with_empty_lists(games)

print(games_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(games['Tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[3:10]:
    print(games.iloc[i[0]].Title)
```

```python
def recommend(game, selected_platform='any'):
    index = games[games['Title']==game].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    list_of_games = []
    for i in distance[3:12]:
        name = games.iloc[i[0]].Title
        game_platform = games.iloc[i[0]].Platforms
        date = games.iloc[i[0]].Release_Date
        plays = games.iloc[i[0]].Plays
        if selected_platform != 'any':
            if name not in list_of_games and selected_platform in game_platform:
                list_of_games.append([name, date, plays])
        else:
            if name not in list_of_games:
                list_of_games.append([name, date, plays])

    for j in list_of_games:
        print(j)
```

```python
recommend('Yume Shoko')
```
